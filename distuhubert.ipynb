```python
# quick_load_test.py  — run this first to confirm the model downloads & runs
import torch
from transformers import AutoFeatureExtractor
from model import DistilHubertClassifier, SAMPLE_RATE

fe = AutoFeatureExtractor.from_pretrained("ntu-spml/distilhubert")
model = DistilHubertClassifier(freeze_backbone=True).eval()

# fake 2 seconds of silence to test the forward pass
import numpy as np
dummy = np.zeros(2 * SAMPLE_RATE, dtype=np.float32)
inputs = fe(dummy, sampling_rate=SAMPLE_RATE, return_tensors="pt")
with torch.no_grad():
    logits = model(inputs["input_values"])
print("OK, logits shape:", logits.shape)  # expect (1, 6)
```
